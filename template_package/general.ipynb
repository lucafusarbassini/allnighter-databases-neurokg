```python
import pandas as pd
df_human = pd.read_parquet('data/liana_humanconsensus_db.parquet')
df_mouse = pd.read_parquet('data/liana_mouseconsensus_db.parquet')
df_human['species'] = "Homo sapiens"
df_mouse['species'] = "Mus musculus"
df = pd.concat([df_human, df_mouse], ignore_index=True)
df
```

```python
collapsed = (
    df.groupby(["source", "target"])["species"]
      .apply(lambda x: sorted(set(x)))   # deduplicate + sort
      .reset_index()
)
collapsed[collapsed['source'] == 'O00182'].head()
```
